```python
z=[0, 0, 1, 0, 0, 0, 1, 0, 0, 2, 3, 1, 1, 0, 1, 1, 0, 1]
z=np.array(z)
z = z.reshape((6, 3))
print(z)
```

```python

from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator

x=np.arange(3)
y=np.arange(6)
X, Y = np.meshgrid(x, y)
Z=z


levels = MaxNLocator(nbins=15).tick_values(z.min(), z.max())
cmap = plt.get_cmap('inferno') #('PiYG')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

fig, ax = plt.subplots(1,1)

img = ax.pcolormesh(X, Y, Z, cmap=cmap, norm=norm)
cbar=fig.colorbar(img, ax=ax, ticks=[0, 0.8, 1.6, 2.4])
cbar.ax.set_yticklabels(['0', '1', '2', '3'], fontsize=12, fontstyle='oblique', color="indigo")
ax.set_title('Colormesh showing numbers of spikes for the full deployment periods for Nodes 17 and 18', fontsize=14, fontstyle='oblique', color="mediumslateblue")
ax.set_yticks((0, 1, 2, 3, 4, 5))
ax.set_yticklabels(("-3", "-2", "-1", "0", "1", "2"), fontsize=12, fontstyle='oblique', color="indigo")
ax.set_xticks((0, 1, 2))
ax.set_xticklabels(("6am-8am", "8am-10am", "10am-13pm"), fontsize=12, fontstyle='oblique', color="indigo")
ax.set_ylabel('Temperature (Degrees C)', fontsize=14, fontstyle='oblique', color="mediumslateblue")
ax.set_xlabel('Time of occurrence', fontsize=14, fontstyle='oblique', color="mediumslateblue")
fig.tight_layout()

plt.show()
```

```python
# contours are *point* based plots, so convert our bound into point
# centers
fig, ax = plt.subplots(1,1)
cf = ax.contourf(X, Y, Z, levels=levels,
                  cmap=cmap)
fig.colorbar(cf, ax=ax, ticks=[0, 1, 2, 3])


ax.set_title('Contourf plot showing numbers of spikes for the period\n16th-30th Jan for Nodes 17 and 18', fontsize=14, fontstyle='oblique', color="indigo")
ax.set_yticks((0, 1, 2, 3, 4, 5))
ax.set_yticklabels(("-3", "-2", "-1", "0", "1", "2"), fontsize=10)
ax.set_xticks((0, 1, 2))
ax.set_xticklabels(("6am-8am", "8am-10am", "10am-13pm"), fontsize=10)
ax.set_ylabel('Temperature (Degrees C)', fontsize=12, fontstyle='oblique', color="indigo")
ax.set_xlabel('Time of occurrence', fontsize=12, fontstyle='oblique', color="indigo")

# adjust spacing between subplots so `ax1` title and `ax0` tick labels
# don't overlap
fig.tight_layout()

plt.show()
```
